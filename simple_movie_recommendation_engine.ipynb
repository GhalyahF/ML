```python
import pandas as pd 
```

```python
#read in files 
user_cols= ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users= pd.read_csv('u.user', sep='|', names= user_cols, encoding='latin-1')

rating_cols=['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings= pd.read_csv('u.data', sep='\t', names=rating_cols, encoding='latin-1')

genre_cols= ['movie_id', 'movie_title', 'release_date', 'video release date', 'IMdb URL', 'Unknown','Action',
             'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime','Documentary', 'Drama', 'Fantasy','Film-noir',
             'Horror', 'Musical', 'Mystery', 'Romance', 'SciFi', 'Thriller', 'War', 'Western']
genres= pd.read_csv('u.item', sep='|', names= genre_cols, encoding='latin-1')
```

```python
ratings
rp= ratings.pivot_table('rating', index='movie_id',columns='user_id')
test=rp[1]

sim_test=rp.corrwith(test)
sim_test
```

```python
rating_c= ratings[test[ratings.movie_id].isnull().values & (ratings.user_id !=1)]
rating_c
```

```python
rating_c['similarity']= rating_c['user_id'].map(sim_test.get)
```

```python
rating_c['sim_rating']= rating_c.similarity * rating_c.rating
rating_c
```

```python
recommendation= rating_c.groupby('movie_id').apply(lambda s: s.sim_rating.sum()/s.similarity.sum())
```

```python
print(recommendation.sort_values(ascending=False))
```
